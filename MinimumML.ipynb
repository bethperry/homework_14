## Linear Regression

Let's predict house prices. Data source [here](https://www.kaggle.com/c/home-data-for-ml-course/data) (for practice, we are just using the train set as our full set)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load in the data
houses = pd.read_csv('data/house_data.csv')
```

```python
# pick your features
pd.set_option('display.max_columns', None)
houses.head(10)
#houses.Neighborhood.unique()
```

```python
houses['Baths'] = houses['FullBath'] + (houses['HalfBath'] * 0.5)
X = houses[['LotArea','GrLivArea','TotRmsAbvGrd','GarageCars','Baths']]
y = houses['SalePrice']
```

```python
#train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
#train it! (don't train on your target column!)
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
#predict it!
regression_model.predict(X_train)
```

```python
#how'd you do?
#accuracy_score(_____, ______)
regression_model.score(X_test,y_test)
```

## Logistic Regression
Let's predict whether a patient has diabetes. Data source [here](https://www.kaggle.com/uciml/pima-indians-diabetes-database)

```python
#load in the data
diabetes = pd.read_csv('data/diabetes_data.csv')
```

```python
#pick your features
print(diabetes.head())
X = diabetes[['Glucose','BMI','Pregnancies']]
y = diabetes['Outcome']
```

```python
#train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
#train it!
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
#predict it!
log_model.predict(X_test)
```

```python
#how'd you do?
log_model.score(X_test, y_test)
```
